```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
health = pd.read_csv(r'/var/health_insurance.csv')
pd.set_option('display.max_rows', 1500)
health
```

```python
health.head()
health['age'].sort_values()

#creating age groups
age_groups = [(17, 35), (36, 45),(46, 55), (56, 65)]

def assign_age_group(age):
    for age_range in age_groups:
        if age_range[0] <= age <= age_range[1]:
            return f"{age_range[0]}-{age_range[1]}"
    return("Unknown")

health['age_group'] = health['age'].apply(assign_age_group)
```

```python
health[['age','age_group']]
```

```python
health['bmi'].min()
health['bmi'].max()

bmi_groups = [(14,18.5), (18.6,24.9), (25,29.9), (30,54)]

def assign_bmi_group(bmi):
    for bmi_range in bmi_groups:
        if bmi_range[0] <= bmi <= bmi_range[1]:
            return f"{bmi_range[0]}-{bmi_range[1]}"
    return("Unknown")

health['bmi_group'] = health['bmi'].apply(assign_bmi_group)
```

```python
health[['bmi','bmi_group']].head()
health['bmi_group'].head()
health.head(10)
```
